```python
data = yf.download(tickers = 'UBER', period = '1d', interval = '1m')
fig = go.Figure()

fig.add_trace(go.Candlestick(x = data.index,
                            open = data['Open'],
                            high = data['High'],
                            low = data['Low'],
                            close = data['Close'],
                            name = 'market data'))

fig.update_layout(
    title = 'uber'
)

fig.update_xaxes(
        rangeslider_visible = True,
        rangeselector = dict(
            buttons = list([
                dict(count = 15, label = '15m', step = 'minute', stepmode = 'backward'),
                dict(count = 45, label = '45m', step = 'minute', stepmode = 'backward'),
                dict(count = 1, label = 'H', step = 'hour', stepmode = 'backward'),
                dict(count = 3, label = '2H', step = 'hour', stepmode = 'backward'),
                dict(step = 'all')
            ])
        )
)
```
